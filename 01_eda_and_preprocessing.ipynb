# 1. Exploratory Data Analysis & Feature Engineering


```python
import pandas as pd
import os
import warnings
warnings.filterwarnings('ignore')

os.makedirs('data/processed', exist_ok=True)
df = pd.read_csv('data/raw/ai4i2020.csv')
print('Original dataset shape:', df.shape)

# 1. Combine Binary Flags to Multiclass Target
def get_failure_type(row):
    for f in ['TWF', 'HDF', 'PWF', 'OSF', 'RNF']:
        if row[f] == 1: return f
    return 'No Failure'

df['Target'] = df.apply(get_failure_type, axis=1)

# 2. Feature Engineering
df['Temperature Diff'] = df['Process temperature [K]'] - df['Air temperature [K]']
df['Power'] = df['Rotational speed [rpm]'] * df['Torque [Nm]'] * 0.10472

# 3. Subset Relevant Features and CLEAN Column Names (XGBoost hates brackets)
features_map = {
    'Air temperature [K]': 'Air temperature',
    'Process temperature [K]': 'Process temperature',
    'Temperature Diff': 'Temperature Diff',
    'Rotational speed [rpm]': 'Rotational speed',
    'Torque [Nm]': 'Torque',
    'Power': 'Power',
    'Tool wear [min]': 'Tool wear',
    'Target': 'Target'
}

processed_df = df[list(features_map.keys())].rename(columns=features_map)

# Save processed dataset
out_path = 'data/processed/processed_data.csv'
processed_df.to_csv(out_path, index=False)
print(f'Saved engineered dataset to {out_path} with shape: {processed_df.shape}')
processed_df.head()
```
